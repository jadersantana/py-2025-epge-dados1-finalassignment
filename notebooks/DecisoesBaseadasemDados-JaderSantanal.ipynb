### Apresentação

Este trabalho tem como objetivo investigar em que medida é possível explicar e prever o comportamento de curto prazo de ações negociadas na B3 a partir de informações historicamente observáveis, como preços passados e volume de negociação.

Os resultados são avaliados não apenas em termos estatísticos (R², acurácia, AUC), mas também sob a ótica econômica, discutindo limitações práticas do uso desses modelos em estratégias de investimento.
